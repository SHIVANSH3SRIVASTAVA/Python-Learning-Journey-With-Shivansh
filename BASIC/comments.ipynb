# Comments In PYTHON

- They are non-executable statements
- They are used to explain code logic, make programs more readable

---

## Types Of Comments In PYTHON

- There are 2 two types of comments in Python
- (1) Single Line Comment (#)
- (2) Multi Line Comments ('''......'')

---

## (1) Single Line Comments

- They are wriiten using "Hash" symbol (#)
- They are used to write single line comments
- You can also use "Ctrl+/" to create a single line comment

```python
# This line is a comment & will be ignored by the interpreter
print("Hello World!")
```

---

## (2) Multi Line Comments

- They are written using triple quotes, either single or double ('''.......''') (""".............""")
- They are used to write multi line comments
- They are also known as "Python Docstrings"

```python
'''
This is a multi line comment & this is 1st line
This is 2nd line of comment
'''

print("Hello World!")
```
