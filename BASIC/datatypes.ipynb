# NUMERIC

## Integer

- Stores integer values

```python
num1 = 5
print(num1)
print(type(num1))
```

---

## Float

- Stores decimal value

```python
num2 = 5.4
print(num2)
print(type(num2))
```

---

## Complex

- Stores complex numbers

```python
num3 = 3+2j
print(num3)
print(type(num3))
```

---

---

# Text

## String

- Used to store string or textual data
- Strings are written inside double inverted comma (" ") or single inverted comma (' ')
- Has a builtin function **str()**

```python
txt = "This is a string"
print(txt)
print(type(txt))
```

---

---

# Sequence

---

## List
- Stores elements of different or same datatypes
- It is ordered
- Use Square brackets [ ] for creating a list
- Elements are separated by comma
- It is mutable
- Has a builtin function **list()**

```python
l = [1,2,"Hello", 8.7, 2+3j]
print(l)
print(type(l))
```

---

## Tuple

- Stores elements of different or same datatypes
- It is ordered
- Use parathesis ( ) to create tuple
- Elements are separated by comma
- It is immutable
- Has a builtin function **tuple()**

```python
t = (1,"Hello",5.6,6+8j)
print(t)
print(type(t))
```

---

## Range
- Used to create sequence of numbers in the given range
- It is immutable
- Mostly used with for loop
- It starts from 0 by default and excludes the stop range
- It is used by calling **range()** function

```python
r = range(6)
print(r)
print(type(r))
print(list(r))
```

# Mapping

## Dictionary

- Used to store data in key-value pair
- Key should be unique
- Values can be duplicate
- Key and Values are separated by colon ( : )
- Elements are separated by comma
- Dictionary is created using curly braces { }
- It is mutable
- Has a builtin function **dict()**

```python
record = {
    "Name" : "Amit",
    "Age" : 35,
    "Marks" : 56.4,
}

print(record)
print(type(record))
```

---

---

# Set

- Use to create set
- It is unordered
- Stores only unique elements
- Use curly braces to create a set { }
- Elements are separated by comma
- It is mutable
- Has a builtin functions **set()**

---

## Normal Set

- Used to create a mutable set

```python
s = {1,3,5,3,2,"Hello",8.2}
print(s)
print(type(s))
```

---

## Frozen Set

- Used to create immutable set
- It uses **frozenset()** function

```python
frozen = frozenset({1,3,5,2,3,1})
print(frozen)
print(type(frozen))
```

---

---

## None
- Use to represent no value

```python
n = None
print(n)
print(type(n))
```
