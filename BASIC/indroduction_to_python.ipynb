# PYTHON

- Python is a high-level, general-purpose programming language celebrated for its exceptional code readability, simplicity, and immense ecosystem
- Python was developed by Guido van Rossum and released in 1991

---

## Features

- High level language
- Interpreted
- Dynamically typed
- Memory management
- Open source
- Simple
- Extensible
- Cross platform supported
- Multi paradigm
- Vast libraries
