## Arithmetic Operator

### 1. Addition

```python
print(2+3)
print(4.5+6) # Return floating point result if any one operand is floating point number
```

## 2. Subtraction

```python
print(9-4)
print(5-9.0) # Return floating point result if any one operand is floating point number
```

## 3. Multiplication

```python
print(3*4)
print(3.0*4) # Return floating point result if any one operand is floating point number
```

## 4. Division
- Return result in floating point number

```python
print(5/2)
print(-5/-2)
print(5.2/2)
print(5/2.0)
```

## 5. Floor Division
- Return integer result

```python
print(-8//2)
print(8.0//2) # Return floating point result if any one operand is floating point number
print(8//2.0) # Return floating point result if any one operand is floating point number
```

## 6. Exponent
- Raises a number to a power

```python
print(2**3)
print(2.0**3) # Return floating point result if any one operand is floating point number
print(-2**3)
```

## 7. Modulus
- Returns remainder
- Sign of the remainder depends on the sign of the denominator

```python
print(7%2)
print(-7%2) # Remainder will be positive because denominator is positive
print(7%-2) # Remainder will be negative because denominator is negative
print(7.0%2) # Return floating point result if any one operand is floating point number
print(7%2.0) # Return floating point result if any one operand is floating point number
```

```python
a=2
b=2
print(a == b) # Compares value
print(a is b) # Compares memory location

x = [1,2,3]
y = [1,2,3]
print(x == y)
print(x is y)
```

```python
a=[1,2,3]
b=[1,2,3]
print(a is not b)

c=a
print(a==b==c)
print(c is a) # Here "c" is now an alias of "a"
print(c is b) # "c" and "b" are two different objects
```

# Bitwise Operator
- Used to perform operations on binary numbers
- Integers are converted into 8 bit format

## 1. AND --> &
- Returns "1" if both bits are "1" else "0"

```python
a = 5 # 00000101
b = 3 # 00000011
print(a & b) # 00000001 = 1

print(1 & 1)
print(1 & 0)
print(0 & 1)
print(0 & 0)
```

---

## 2. OR --> |
- Returns "1" if either of the bits are "1"

```python
a = 5 # 00000101
b = 3 # 00000011
print(a | b) # 00000111 = 7

print(1 | 1)
print(1 | 0)
print(0 | 1)
print(0 | 0)
```

---

## 3. NOT --> ~
- It flips every bit (0 -> 1) (1 -> 0)
- In Python, integers are represented using 2's complement => ~n = -(n+1)

```python
a = 5 # 00000101
print(~a) # 11111010 --> -(5+1) = -6

print(~1) # 00000001 -> 11111110 = -2
print(~0) # 00000000 -> 11111111 = -1
```

---

## 4. XOR --> ^
- Returns "1" for different bits otherwise "0"
- Same "0" different "1"

```python
a = 5        # 00000101
b = 3        # 00000011
print(a ^ b) # 00000110 = 6

print(1 ^ 1)
print(1 ^ 0)
print(0 ^ 1)
print(0 ^ 0)
```

---

## 5. Left Shift --> <<
- Multiplies by power of 2 by adding 0's to the rightmost side
- FORMULA --> **a << n = a x 2^n**

```python
a = 5         # 0101
print(a << 1) # 01010 = 10  (Multiplied by 2^1)
print(a << 2) # 010100 = 20 (Multiplied by 2^2)
```

---

## 6. Right Shift --> >>
- Divides by power of 2 by removing the bit of the rightmost side and adding same no. of bits on leftmost side
- FORMULA --> **a >> n = a ÷ 2^n**

```python
a = 20        # 10100
print(a >> 1) # 01010 = 10 (Divide by 2^1)
print(a >> 2) # 00101 = 5  (Divide by 2^2)
```
