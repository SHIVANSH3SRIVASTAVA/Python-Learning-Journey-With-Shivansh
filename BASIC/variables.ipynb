# Variables

- Variables are containers that are used to store some values or data

## Naming Conventions

- Do not start with numbers
- Case sensitive
- No space between
- No special characters other than underscore (_)
- Should not be a keyword
- Start with small letter

### Examples

```python
num_1 = 53
print(num_1)
```

```python
name = "Shivansh Srivastava"
print(name)
```

- In the above examples, (num_1) and (name) are variables and (53) and (Shivansh Srivastava) are the data stored in that variables
