# Decision Making In Python
- They are used to perform an action on the basis of certain conditions

## Types Of Decision Making Statements In Python
- There are three types of decision making statements in Python
- There can be any number of conditions with the decision making statements

1. If
2. Elif
3. Else

---

# 1. If
- Used to perform an action if the given condition is True

## SYNTAX

### Examples

```python
a = 10
if(a == 10):
    print("If statement executed")
```

---

## 2. Elif
- Elif condition is checked when if condition is False

## SYNTAX

## Examples

```python
a = 15
if(a<10):
    print("If is executed")
elif(a>10):
    print("Elif is executed")
```

---

## 3. Else
- Else is executed if neither if nor elif conditions is True
- There is no condition with else

## SYNTAX

## Examples

```python
age = 18
if(age<18):
    print("Age is smaller")
elif(age>18):
    print("Age is bigger")
else:
    print("You are eligible")
```
